```python
X = dataset[["tickets_sold","ticket_use","show_time","occu_perc","capacity","ticket_price","tickets_out"]]
y = dataset["total_sales"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import xgboost as xgb
# Initialize XGBoost regressor
xgb_reg = xgb.XGBRegressor(objective='reg:squarederror')

# Train the model
xgb_reg.fit(X_train, y_train)
```

```python
y_pred1 = xgb_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_squared = r2_score(y_test, y_pred1)
r_squared
```

```python
# Assuming y_pred and y_test are NumPy arrays
result_df2 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred1})
```

```python
import gradio as gr

def predict(tickets_sold: int,
             ticket_use: int,
             show_time: int,
             occu_perc: float,
             capacity: float,
             ticket_price: float,
             tickets_out: int):

    input_data = np.array([[tickets_sold, ticket_use, show_time, occu_perc, capacity, ticket_price, tickets_out]]).astype(np.float64)
    input_data = sc.transform(input_data)
    prediction = xgb_reg.predict(input_data)
    return prediction

iface = gr.Interface(predict,
                     inputs=[gr.Number(label="tickets_sold"),
                             gr.Number(label="ticket_use"),
                             gr.Number(label="show_time"),
                             gr.Number(label="occu_perc"),
                             gr.Number(label="capacity"),
                             gr.Number(label="ticket_price"),
                             gr.Number(label="tickets_out")],
                     outputs="number")

iface.launch(share=True)
```
